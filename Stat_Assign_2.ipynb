### 1. Mean vs. Median

**Mean:**
- **Definition:** The mean (or average) is the sum of all the values in a dataset divided by the number of values.
- **Formula:** 
  \[
  \text{Mean} = \frac{\sum_{i=1}^{N} X_i}{N}
  \]
  where \( X_i \) are the individual data points and \( N \) is the number of data points.
- **Usage:** The mean is used to measure the central tendency of a dataset.
- **Sensitivity to Outliers:** The mean is sensitive to outliers, which can skew the average if there are extreme values in the dataset.
- **Example:**
  - Dataset: [1, 2, 3, 4, 5]
  - Mean: \( \frac{1+2+3+4+5}{5} = 3 \)

**Median:**
- **Definition:** The median is the middle value in a dataset when it is ordered from smallest to largest. If the dataset has an even number of values, the median is the average of the two middle values.
- **Usage:** The median is used to measure the central tendency, especially for skewed distributions or datasets with outliers.
- **Robustness:** The median is robust to outliers and provides a better measure of central tendency for skewed distributions.
- **Example:**
  - Dataset: [1, 2, 3, 4, 5]
  - Median: 3 (middle value)
  - Dataset with an even number of values: [1, 2, 3, 4, 5, 6]
  - Median: \( \frac{3+4}{2} = 3.5 \)

### 2. Mode Analysis in Distribution

**Mode:**
- **Definition:** The mode is the value that appears most frequently in a dataset.
- **Usage:** The mode is used to identify the most common value in a dataset and can be useful for categorical data.
- **Types of Mode:**
  - **Unimodal:** A distribution with a single mode.
  - **Bimodal:** A distribution with two modes.
  - **Multimodal:** A distribution with more than two modes.
- **Example:**
  - Dataset: [1, 2, 2, 3, 4]
  - Mode: 2 (most frequent value)
- **Analysis in Distribution:**
  - The mode can help identify the peak(s) in a distribution.
  - In a normal distribution, the mode is the same as the mean and median.
  - In skewed distributions, the mode can be different from the mean and median.

### 3. Types of Distributions

**1. Normal Distribution:**
- **Characteristics:** Symmetrical, bell-shaped curve. Mean, median, and mode are all equal and located at the center.
- **Example:** Heights, test scores.

**2. Uniform Distribution:**
- **Characteristics:** All values have the same probability of occurring. The distribution is rectangular in shape.
- **Example:** Rolling a fair die.

**3. Binomial Distribution:**
- **Characteristics:** Discrete distribution representing the number of successes in a fixed number of independent Bernoulli trials.
- **Example:** Number of heads in 10 coin flips.

**4. Poisson Distribution:**
- **Characteristics:** Discrete distribution representing the number of events occurring in a fixed interval of time or space.
- **Example:** Number of emails received in an hour.

**5. Exponential Distribution:**
- **Characteristics:** Continuous distribution representing the time between events in a Poisson process.
- **Example:** Time between arrivals of customers at a store.

**6. Skewed Distributions:**
- **Left-Skewed (Negative Skew):** Tail is on the left side. Mean < Median < Mode.
  - **Example:** Age at retirement.
- **Right-Skewed (Positive Skew):** Tail is on the right side. Mode < Median < Mean.
  - **Example:** Income distribution.

**7. Log-Normal Distribution:**
- **Characteristics:** If a variable's logarithm is normally distributed, the variable is log-normally distributed.
- **Example:** Stock prices.

**8. Student's t-Distribution:**
- **Characteristics:** Similar to normal distribution but with heavier tails. Used when sample size is small and population standard deviation is unknown.
- **Example:** Small sample hypothesis testing.

**9. Chi-Square Distribution:**
- **Characteristics:** Used in hypothesis testing and confidence interval estimation for variance.
- **Example:** Chi-square tests for independence.
